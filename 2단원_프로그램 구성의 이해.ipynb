**단원 열기**  
우리는 살아가면서 많은 선택을 한다. 쇼핑을 하더라도 A가 
더 좋은 제품인지, B가 더 좋은 제품인지 여러 가지 조건을 비교
해가며 선택한다. 또한, 다이어트처럼 목표를 정하고 그 목표가 달성
될 때까지 일련의 과정(운동 및 식습관)들을 반복해야 하는 경우도 있다. 
또한, 일정한 패턴이 있는 업무일 경우 매뉴얼을 만들고 해당 매뉴얼대로 처
리하거나 다른 사람에게 해당 업무를 맡기고 결과를 보고 받는다.  
컴퓨터도 마찬가지다. 조건문을 사용하여 특정 조건을 만족했을 때 해야 할 일을 정의
해줘야 하고, 특정 조건을 만족할 때까지 반복하여 처리해야 할 경우 반복문을 사용하여 
처리해야 한다. 또한, 일정한 패턴이 반복되는 일인 경우 함수로 만들어 필요할 때마다 함수
를 호출하여 코드를 재사용하고 불필요한 코드 반복이 일어나지 않도록 하는 것이 중요하다.

````python
# 예2-23: 리턴할 값이 많은 경우(1)
def sum_avg_a_to_b(a , b):
    if(a < 0 or b < 0):
        print("입력한 값이 자연수가 아닙니다.")
        return(None, None) # 에러처리, 리턴할 값이 존재하지 않아 None 객체 반환
    if(a > b):
        c=b
        b=a
        a=b
    _sum=0
    j=0 # 평균을 구하기 위한 a와 b 사이의 숫자 카운트
    for i in range(a, b+1, 1) :
        _sum=_sum+i
        j=j+1
    _avg=_sum/j
    return (_sum, _avg)

while(True):
    print("a부터 b까지 합계를 구합니다.")
    print("프로그램 종료: a=0, b=0 입력")
    num1 = int(input("첫 번째 자연수를 입력하세요."))
    num2 = int(input("두 번째 자연수를 입력하세요."))
    if(num1==0 and num2==0):
        break
    ret_sum, ret_avg=sum_avg_a_to_b(num1 , num2)
    if(ret_sum != None or ret_avg != None):
        print(num1,"부터 ", num2 ,"까지 합계: ", ret_sum)
        print(num1,"부터 ", num2 ,"까지 평균: ", ret_avg)
print("프로그램을 종료합니다.")
````

````python
# 예2-25: 내장 함수로 대체했을 때(1)
def sum_avg_a_to_b(a , b):
    if(a < 0 or b < 0): # 입력값이 0보다 작은 경우
        print("입력한 값이 자연수가 아닙니다.")
        return(None, None) # 에러처리, 리턴할 값이 존재하지 않아 None 객체 반환
    if(a > b): # 작은 수가 a, 큰 수가 b가 되도록 교환
        c=b
        b=a
        a=b  
    _sum=0 # 값 초기화
    j=0 # 값 초기화
    for i in range(a, b+1, 1) : # a~b까지 합계
        _sum=_sum + i # 더한 값을 누적
        j=j+1 # a와 b 사이의 숫자 개수 카운트
    _avg=_sum/j # 평균 구하기
    return (_sum, _avg) #튜플로 2개의 값 리턴

while(True):
    print("a부터 b까지 합계를 구합니다.")
    print("프로그램 종료: a=0, b=0 입력")
    num1 = int(input("첫 번째 자연수를 입력하세요."))
    num2 = int(input("두 번째 자연수를 입력하세요."))
    if(num1==0 and num2==0):
        break
    ret_sum, ret_avg=sum_avg_a_to_b(num1 , num2)
    if(ret_sum != None or ret_avg != None):
        print(num1,"부터 ", num2 ,"까지 합계: ", ret_sum)
        print(num1,"부터 ", num2 ,"까지 평균: ", ret_avg)

print("프로그램을 종료합니다.")
````

````python
# 예2-26: 내장 함수로 대체했을 때(2)
def sum_avg_a_to_b(a , b):
    if(a < 0 or b < 0): # 입력값이 0보다 작은 경우
        print("입력한 값이 자연수가 아닙니다.")
        return(None, None) # 에러처리, 리턴할 값이 존재하지 않아 None 객체 반환
    if(a > b): # 작은 수가 a, 큰 수가 b가 되도록 교환
        c=b
        b=a
        a=b  
    _sum=sum(range(a, b+1)) # a~b까지 합계
    _avg=_sum/len(range(a, b+1)) # 평균 구하기
    return (_sum, _avg) # 튜플로 2개의 값 리턴

while(True):
    print("a부터 b까지 합계를 구합니다.")
    print("프로그램 종료: a=0, b=0 입력")
    num1 = int(input("첫 번째 자연수를 입력하세요."))
    num2 = int(input("두 번째 자연수를 입력하세요."))
    if(num1==0 and num2==0):
        break
    ret_sum, ret_avg=sum_avg_a_to_b(num1 , num2)
    if(ret_sum != None or ret_avg != None):
        print(num1,"부터 ", num2 ,"까지 합계: ", ret_sum)
        print(num1,"부터 ", num2 ,"까지 평균: ", ret_avg)

print("프로그램을 종료합니다.")
````

````python
# 예2-30: 수학 관련 함수
arr=[4,9,10,2,8]
print(sorted(arr)) # 정렬된 새로운 리스트를 반환함
print(arr) # arr 리스트 자체를 정렬하지 않았으므로 초기 상태 유지
print(list(reversed(arr))) # reversed()의 리턴 객체를 리스트로 변환 출력 
print(type(reversed(arr))) # reversed()의 리턴 객체의 타입 확인
print(tuple(reversed(arr)))# 튜플형으로 변환
a,b,c,d,e=(reversed(arr)) # 튜플형으로 변환하여 각 변수에 저장
print(a,b,c,d,e)
````

````python
# 예2-30: map 함수 (1)
def square(num):
    return(num*num)
num_list=[1,2,3,4,5]
for i in range(len(num_list)) :
    num_list[i] = square(num_list[i])

print(num_list)
````

````python
# 예2-30: map 함수 (2)
def square(num):
    return(num*num)

num_list=[1,2,3,4,5]
print(list(map(square, num_list)))
````
