
# 타이타닉 데이터

## 문제
다음은 캐글에서 제공하는 ‘타이타닉’ 데이터를 불러와서 분석하는 프로그램의 일부이다. 주석의 내용이 출력되도록 빈칸을 완성해보자.
```
• survived : 생존=1, 죽음=0
• pclass : 승객 등급. 1등급=1, 2등급=2, 3등급=3
• sibsp : 함께 탑승한 형제 또는 배우자 수
• parch : 함께 탑승한 부모 또는 자녀 수
• ticket : 티켓 번호
• cabin : 선실 번호
• embarked : 탑승 장소. S=Southhampton, C=Cherbourg, Q=Queenstown
```

## 주의
반드시 print() 함수를 사용하여 해당 데이터를 출력해주세요.

````python
#답안
import seaborn as sns
df = sns.load_dataset('titanic') 

# 데이터의 첫 7줄 출력
print(df.head(7))

# 빈 데이터의 칼럼별 합계 확인
print(df.isnull().sum() )

# 생존 여부(survived)의 값별 인원 확인
print(df['survived'].value_counts())

# 성별(sex)에 따른 생존여부(survived)의 평균(=생존율)
print(df.groupby('sex')['survived'].mean())

# 승객등급(pclass)에 따른 생존여부(survived)의 평균(=생존율)
print(df.groupby('pclass')['survived'].mean())

# 성별(sex), 승객등급(pclass)에 따른 생존여부(survived)의 평균(=생존율)
print(df.groupby(['sex', 'pclass'])['survived'].mean())
````
